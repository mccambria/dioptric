```python
import serial

ser = serial.Serial(
    port="COM6",  # The COM port your device is connected to
    baudrate=9600,  # Baud rate (e.g., 9600, 115200) - must match your device
    bytesize=8,  # Number of data bits (e.g., 8)
    parity="N",  # Parity (N for None, E for Even, O for Odd)
    stopbits=1,  # Number of stop bits (e.g., 1)
    timeout=1,  # Read timeout in seconds (optional)
)
```

```python
ser.write("0bw".encode("utf-8"))
response = ser.readline().decode("utf-8")
```
